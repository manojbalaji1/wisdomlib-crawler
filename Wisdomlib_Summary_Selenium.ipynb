```python
from selenium.webdriver.common.by import By
import os
driver = webdriver.Chrome(options=chrome_options)
not_done=[]
landing_page_url = 'https://www.wisdomlib.org/hinduism/book/the-skanda-purana/'
driver.get(landing_page_url)
a_tags = driver.find_elements(By.TAG_NAME, 'a')

# urls=[x.get_attribute('href') for x in a_tags]
# print(urls)
url_elements = driver.find_elements("css selector", '#indexList > div > div.il-holder > div > span > a')
urls = [url_element.get_attribute('href') for url_element in url_elements]
urls = urls[:26]
print(url_elements)

for url in urls:
    print(url)
    driver.get(url)
    url_elements = driver.find_elements("xpath", '//span[contains(@class, "il-sign")]//a')
    urls_book = [url_element.get_attribute('href') for url_element in url_elements]
    book_name = driver.find_element("css selector", "#pageContent > div > section.col-12.heading > div > div > nav > span > a").text
    file_path = f'/content/drive/MyDrive/UCB-Mitra/wisdomlib-dataset/english/{book_name}'
    os.makedirs(file_path, exist_ok=True)
    section_name = driver.find_element("css selector", "#pageContent > div > section.col-12.heading > div > h1").text
    print(book_name, "    ", section_name)
    file_path = file_path+"/"+section_name
    os.makedirs(file_path, exist_ok=True)
    print(urls_book)
    for book_url in urls_book:
        print(book_url)
        driver.get(book_url)
        url_elements = driver.find_element("id", 'scontent')
        fname=driver.find_element("xpath","//h1[contains(@class,'pt-2')]").text
        with open(f"{file_path}/{fname}.txt","w") as file:
            file.write(url_elements.text)


    # # Copy the desired text from the page using appropriate methods
    # text_element = driver.find_element_by_xpath('//div[@class="text-class"]')
    # copied_text = text_element.text

    # # Do something with the copied text, e.g., store it or process it

    # # Click on the "Next" button or navigate to the next URL
    # next_button = driver.find_element_by_xpath('//button[@id="next-button"]')
    # next_button.click()
    # driver.quit()

driver.quit()
```
